## Neural Stimulation

The primary way the MEA allows us to interface with an organoid is by sending electrical signals to training neurons that are in said organoid. When the neurons receive these electrical signals it causes them to fire. That is why this process is called **Neural Stimulation**. Making neurons fire over and over again at specific times can permanently change the firing pattern of the neuron. This process is called **Neural Plasticity** or **Synaptic Plasticity**.

### What is a Stimulation Pulse?

A stimulation pulse occurs when we use one of the electrodes on the MEA to change the voltage for a short period of time. This change in voltage only affects neurons in the immediate vicinity of the voltage change. Below is a plot of the most common type of stimulation pulse we use for experiments. This type of pulse is called a **square wave**. We raise the voltage 150mv and then lower it by 150mv, bringing the voltage back to the resting state. In total, this pulse takes 200 μs seconds from start to finish. Note that μs means microseconds, where there are 1000 microseconds in a millisecond. Notice that this pulse resembles the change in mV seen when a  [neuron spikes](https://en.wikipedia.org/wiki/Action_potential).

When creating stimulation protocols, we usually try to avoid protocols with long high voltage pulses or very high frequencies. This will cause the neuron firing in the organoid to tire out, this will not only make the organoid "sluggish", but might also damage it. Because of this we don't typically use voltages over 150 mV and frequencies above 30 Hz, nor overly long stim phases. Generally we want to stay around 10 Hz, but training signals can go higher.

## Experiment Protocol Explained

The experiment who's data we have provided for you was performed on an organoid and recorded from an MEA, it was a basic cart-pole experiment, like the ones we had you do in the last class, where the organoid was tasked with balancing a pole on a moving cart in a simulated environment. For a more in depth discussion of how the experiment was set up and ran please refer to this paper : [Goal-Directed Learning in Cortical Organoids](https://pubmed.ncbi.nlm.nih.gov/39713376/), however we will provide a basic summary here:

There are three types of neurons we can stimulate and read activity from: (1) there are two input neurons that decode the signal we send them about pole position, (2) there are 2 output neurons whose activity encodes how the organoid chooses to push the cart, and (3) there are _n_=6 training neurons in the experiment given to you.

- <b>Encoding:</b> We used rate-encoding to send the computer side state of the pole, on two neurons we stimulate based on the pole angle $\theta$: $N_1 = a\cdot (b-\sin(\theta))^2$ and $N_2 = a\cdot (b+\sin(\theta))^2$ with $a=7$ and $b=0.15$ as a scaling factor to keep stimulation frequencies in a biologically relevant range. Essentially, as the pole moves left, one neuron increases stimulation frequency while the other decreases, and the other way around for when the pole leans right.

- <b>Decoding:</b> Like with encoding we used rate-decoding on two neurons. We use a threshold to detect significant spikes, and then activity is converted to a smooth firing rate with the following function $r_t=\alpha r_{t-1}+(1-\alpha)c_t$, where $r_t$ is the smoothed firing rate at time $t$, $c_t$ is the raw spike count in the current window, and $\alpha=0.2$ balances immediate responses with firing rate stability. The force applied to the cart below the pole is then determined by the difference between the two decoding neuron's smoothed activity (clipped between $-1$ and $1$) and then multiplied by $10$. 

**Training:** For each training neuron we use a chosen single electrode to train it. These electrodes are indexed by $i$: $1 \leq i\leq n$. The training neurons are part of a neural circuit connecting the input neurons to the output neurons.

 We borrow ideas from reinforcement learning to train the training neurons. Let's say an episode of play lasts until the organoid drops the pole and that a session of play consists of 5 episodes. First, we define a notion of numerical **reward** that reflects the task performance; the number of timesteps the pole balanced. The exact reward is calculated for a 5-episode session of play relative to the performance in a 20-episode session containing that 5-episode session to determine when the organoid is not currently improving. We deliver a training pulse at this time to modify the organoids current state.
 
 The training pulses are optimized with the temporal difference learning algorithm, with a large learning rate to quickly adapt. This helps learn $V_i$. External value learning is necessary because the organoid does not itself have the rest of the neural machinery that is there in an actual animal to administer rewards to itself, e.g. dopamine signaling, etc., and learn state values. The external value learning algorithm serves as a replacement for this, modifying activity whenever it is not improving, and learning which training stimulations do this best.
 Learning occurs over many sessions of play. For session $t$, the reward is $R_t$. The expected reward for firing neuron $i$ that was estimated at the time of session $t$ is $V_{i,t}$.  The learning algorithm compares the actual reward $R_t$. to the expected reward $V_{i,t}$. If the actual reward is higher, it sets $V_{i,t+1}$, the expected reward for firing neuron $i$ in the next session, to a higher value than $V_{i,t}$, else if the actual reward is lower, it sets it to a lower value. The formula used for this is $V_{i,t+1} = V_{i,t}, + \alpha(R_t - V_{i,t})E_{i,t}$. Here the positive constant $\alpha$ = 0.3 is a learning rate. The number  $E_{i,t}$ is an abstraction of what we discussed in class as an eligibility trace. In biology, an eligibility trace serves as a kind of "chemical memory" in a neuron summarizing recent activity, so that when a reward signal comes, the change to the neuron is based not just on whether it is firing at the moment (like in a pure Hebb rule), but on its recent firing history. Such a chemical memory decays with time at a certain rate gamma if the neuron is not fired. This is where the gamma-discounting comes in for the pole-balancing training. Instead of gamma-discounting future rewards, we gamma discount past reward activity of the neuron when calculating the value change for the current reward.  The formula is  $E_{i,t} = \gamma E_{i,t-1} + 1$ if neuron $i$ is getting a reinforcement pulse at time $t$, else $E_{i,t} = \gamma E_{i,t-1}$.

Neurons are reinforced in pairs, one right after the other, rather than one neuron at a time, as described in the paper. This is to promote changes to the synaptic strengths between specific pairs of pre- and post-synaptic neurons, rather than changing all synapses of a single post-synaptic neuron equally, which is what you might get if each neuron was pulsed individually. We don't know the precise synaptic connections between neurons, so there is considerable guesswork in choosing pairs of neurons to stimulate. A simple heuristic for this is used, as described in the paper.  The nature of the pulse signals that are used to administer reinforcement was determined from prior experimentation, and is described in the paper. More research can and will be done here too. Finally, the pluses are delivered when performance is below what is desired, so in a way they could be considered punishment (preferably training) rather than reward.

## STTC

## Causal Connectivity

However, you may notice the STTC matrix is symmetrical, this means that it does not show directionality. It is still a useful metric, however we can get more info on how information flows with something called a Causal Connectivity Matrix. Causal Connectivity ues latencies to let us to see the one way correlation between two neural units, i.e. the rate of neural unit A firing after neural unit B is different from the rate of B firing after A. Below is are the causal connectivity graphs for your organoid, specifically for the encode(first two), decode(next two), and training neurons(remaining 6):

```python
causal_plot()
```


What is really useful about Causal Connectivity is we can see how the neural units chain together. When looking directly at the matrix the first thing that will be obvious is what neural units fire directly after anothers; these show a one-way connection but it goes further than that. But we can tell the order of the connections. If something is a first order connection there is a direct link between the two units, if A and B are in an nth order connection that means there are n-1 other neural units relaying the info from A before it reaches B, there are n-1 units "between'' A and B. Using the CC matrix on different time intervals we can distinguish the order of connections. The two matrices above are taken over two scales, the first is taken over 10-15ms range, the second is taken over a pretty large time frame (200ms). From these we can see that neuron 2 is very unreactive in the first order but reacts to all inputs in Nth order.

## Latencies

Using the STTC we learned a way to measure the degree to which neurons fire together. But is there actually any degree of communication between the neurons? One way to parse out this question is to see if one neuron consistently fires before the other. This is done by using a measurement called **spike time latency**, or latencies for short. 

Suppose we want to calculate the latency between some spike on a neuron ($n_1$) compared to another neuron ($n_2$). The latency is calculated by measuring the duration of time in between the spike of $n_1$ and the closest occurring spike from $n_2$. If the $n_1$ spike occurs before the $n_2$ spike this distance is positive, otherwise, the distance is negative. Doing this for all the spikes of $n_1$ we can create a histogram of the latencies between two neurons. if the histogram has mostly positive values, or mostly negative values, then we have a general notion of the direction of "information flow".

The interactive graphic below plots a histogram of the latencies between any two neural units. Try running the widget using the dataset `baseline` with $n_1=48$ and $n_2=38$. Notice that these two neurons that are located close to each other have latencies that are skewed negatively. Now make $n_1=38$ and $n_2=48$. Notice that the latencies are now skewed positively.

## Principal Component Analysis & Eigenvector Mapping

In a previous homework, we delved into the subject of the correlation and Spike Time Tiling Coefficient matrices. These can be useful tools to understanding the relationships of the neurons in the culture, specifically how they act together in unison. However when looking at these matrices It can be somewhat difficult to glean specific structures from them, but them being matrices, diagonal matrices to be specific, allows us to harness an important aspect of linear algebra to simplify their information. 

As discussed in class, the eigenvectors of a matrix represent the axes upon which the matrix's transform operates, with the corresponding eigenvalue being the degree to which that transform is performed. The eigenvectors span a vector space known as the eigenspace, the subspace where the matrix  affects another vector, by stretching or compressing the vector with respect to the matrix's eigenvalues. This means certain axes, therefore eigenvectors, contribute more to the transformation then others, since their eigenvalues affect the system to a greater degree. From this we can decompose any diagonalizable matrix into its set of eigenvectors, select the eigenvectors that affect this space the most and reconstruct a pretty close approximation to the matrix using only those, and therefore fewer, elements. This simplification allows us to understand the matrix through a lower dimension. Because our correlation and STTC matrices are symmetric, we can easily get their eigenvalues and eigenvectors with the following functions.


```python
Corr = correlation(sd_main)
STTC = get_sttc(sd_main)

Wcorr, Ucorr = eigenvalues_eigenvectors(Corr)
Wsttc, Usttc = eigenvalues_eigenvectors(STTC)

EigenvectorAnalysis(Ucorr, Usttc)
```
